# Ödev 3: Özdeğer ve Özvektör Hesaplama ve Görsel Analizi

Bu çalışma, makine öğrenmesinin temelini oluşturan özdeğer (eigenvalue) ve özvektör (eigenvector) kavramlarını hem teorik hem de uygulama düzeyinde incelemektedir.

### Çalışma Kapsamı:
1. **Numpy linalg.eig** ile profesyonel hesaplama.
2. **Karakteristik Denklem** ile manuel (analitik) doğrulama.
3. Özvektörlerin **geometrik dönüşüm** üzerindeki etkisinin görselleştirilmesi.
4. Özdeğerlerin **Varyans (PCA)** ile ilişkisinin analizi.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Analiz Edilecek Matrisin Tanımlanması
# Örnek: A = [[6, 2], [1, 5]]
A = np.array([[6, 2], [1, 5]])

print("--- İncelenen Matris ---")
print(A)
print("-----------------------")
```

```python
# Yöntem 1: Numpy linalg.eig Kullanımı
eig_vals_np, eig_vecs_np = np.linalg.eig(A)

# Yöntem 2: Manuel (Analitik) Hesaplama
# Karakteristik Denklem: det(A - λI) = 0
# (6-λ)(5-λ) - (2*1) = 0  => λ^2 - 11λ + 28 = 0
a, b, c = 1, -11, 28
delta = b**2 - 4*a*c
lambda_1 = (-b + np.sqrt(delta)) / (2*a)
lambda_2 = (-b - np.sqrt(delta)) / (2*a)
eig_vals_manual = np.array([lambda_1, lambda_2])

print(f"Numpy Özdeğer Sonuçları: {np.sort(eig_vals_np)}")
print(f"Manuel Hesaplama Sonuçları: {np.sort(eig_vals_manual)}")

if np.allclose(np.sort(eig_vals_np), np.sort(eig_vals_manual)):
    print("\n✅ SONUÇ: Hesaplama yöntemleri birbiriyle tam uyumludur.")
```

## Geometrik Görselleştirme
Özvektörler, bir matris dönüşümü uygulandığında doğrultusu değişmeyen özel vektörlerdir. Aşağıdaki grafik, matrisin bir birim daireyi nasıl elipse dönüştürdüğünü ve özvektörlerin bu yapı üzerindeki konumunu göstermektedir.

```python
def plot_transformation(A):
    vals, vecs = np.linalg.eig(A)
    theta = np.linspace(0, 2*np.pi, 100)
    circle = np.array([np.cos(theta), np.sin(theta)])
    transformed = A @ circle

    plt.figure(figsize=(8, 8))
    plt.plot(circle[0], circle[1], '--', color='gray', label='Orijinal (Birim Daire)')
    plt.plot(transformed[0], transformed[1], color='blue', label='Dönüştürülmüş Veri')

    colors = ['red', 'green']
    for i in range(len(vals)):
        v = vecs[:, i] * vals[i]
        plt.quiver(0, 0, v[0], v[1], color=colors[i], angles='xy', 
                   scale_units='xy', scale=1, label=f'Özvektör {i+1} (λ={vals[i]:.1f})')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.title("Lineer Dönüşüm ve Özvektörlerin Etkisi")
    plt.axis('equal')
    plt.show()

plot_transformation(A)
```

## Makine Öğrenmesi Uygulaması: Bilgi Miktarı Analizi
Özdeğerlerin toplamı içindeki payı, o özvektörün verideki varyansın (bilginin) ne kadarını taşıdığını gösterir. Bu prensip **PCA (Temel Bileşen Analizi)** algoritmasının temelidir.

```python
total_eig = np.sum(eig_vals_np)
for i, val in enumerate(eig_vals_np):
    importance = (val / total_eig) * 100
    print(f"Bileşen {i+1}: %{importance:.2f} oranında varyans (bilgi) taşımaktadır.")
```
