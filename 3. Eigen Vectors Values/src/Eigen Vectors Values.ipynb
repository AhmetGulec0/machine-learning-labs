# Ödev 3: Özdeğer ve Özvektör Hesaplama Analizi

Bu notebook kapsamında **Numpy linalg.eig** ve **Manuel (Analitik)** yöntemler karşılaştırılmıştır.

```python
import numpy as np

# Matris Tanımı
A = np.array([[6, 2], [1, 5]])

# 1. Numpy Yöntemi
eig_vals_np, _ = np.linalg.eig(A)

# 2. Manuel Yöntem (Karakteristik Denklem çözümü)
# λ^2 - 11λ + 28 = 0
a, b, c = 1, -11, 28
delta = b**2 - 4*a*c
lambda_1 = (-b + np.sqrt(delta)) / (2*a)
lambda_2 = (-b - np.sqrt(delta)) / (2*a)
eig_vals_manual = [lambda_1, lambda_2]

print(f"Numpy Sonuçları: {np.sort(eig_vals_np)}")
print(f"Manuel Sonuçlar: {np.sort(eig_vals_manual)}")
```
