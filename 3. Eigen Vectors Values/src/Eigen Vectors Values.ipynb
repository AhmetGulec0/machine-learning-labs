# Ödev 3: Özdeğer ve Özvektör Hesaplama Analizi

Bu çalışma kapsamında, bir matrisin özdeğerleri hem **Numpy** kütüphanesi ile hem de **Manuel (Analitik)** yöntemle hesaplanarak karşılaştırılmıştır. Ayrıca özvektörlerin geometrik anlamı görselleştirilmiştir.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Analiz Edilecek Matrisin Tanımlanması
A = np.array([[6, 2], [1, 5]])

# 2. Yöntem: Numpy linalg.eig Kullanımı
eig_vals_np, eig_vecs_np = np.linalg.eig(A)

# 3. Yöntem: Manuel (Analitik) Hesaplama
# Karakteristik Denklem: det(A - λI) = 0
# λ^2 - 11λ + 28 = 0
a, b, c = 1, -11, 28
delta = b**2 - 4*a*c
lambda_1 = (-b + np.sqrt(delta)) / (2*a)
lambda_2 = (-b - np.sqrt(delta)) / (2*a)
eig_vals_manual = np.array([lambda_1, lambda_2])

print("--- HESAPLAMA SONUÇLARI ---")
print(f"Numpy Özdeğerleri: {np.sort(eig_vals_np)}")
print(f"Manuel Özdeğerler: {np.sort(eig_vals_manual)}")
print("---------------------------")
```

## Özvektörlerin Geometrik Görselleştirilmesi
Aşağıdaki grafik, matris dönüşümü sonucunda yönü değişmeyen (sadece ölçeklenen) özvektörleri göstermektedir.

```python
def plot_eigenvectors(A):
    vals, vecs = np.linalg.eig(A)
    origin = [0], [0]
    
    plt.figure(figsize=(7, 7))
    
    # Birim daire ve dönüşümü
    theta = np.linspace(0, 2*np.pi, 100)
    circle = np.array([np.cos(theta), np.sin(theta)])
    transformed_circle = A @ circle
    
    plt.plot(circle[0], circle[1], color='gray', linestyle='--', label='Birim Daire')
    plt.plot(transformed_circle[0], transformed_circle[1], color='blue', label='Dönüşmüş Veri')
    
    # Özvektörlerin çizimi
    colors = ['red', 'green']
    for i in range(len(vals)):
        v = vecs[:, i] * vals[i]
        plt.quiver(*origin, v[0], v[1], color=colors[i], 
                   angles='xy', scale_units='xy', scale=1, 
                   label=f'Özvektör {i+1} (λ={vals[i]:.1f})')

    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True)
    plt.legend()
    plt.title("Matris Dönüşümü ve Özvektörler")
    plt.show()

plot_eigenvectors(A)
```

## Uygulama: Varyans ve Özdeğer İlişkisi
Özdeğerlerin büyüklüğü, o yöndeki verinin varyansını (bilgi miktarını) temsil eder. Bu durum PCA algoritmasının temelidir.

```python
for i in range(len(eig_vals_np)):
    variance_ratio = eig_vals_np[i] / np.sum(eig_vals_np) * 100
    print(f"Özdeğer {eig_vals_np[i]}: Verideki bilgi miktarının %{variance_ratio:.2f}'sini temsil eder.")
```
